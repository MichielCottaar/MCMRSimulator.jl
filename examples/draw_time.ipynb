```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randn(100, 2)
position = np.cumsum(x, 0)
displacement = np.linalg.norm(position, axis=-1)
plt.plot(*position.T)
```

```python
plt.plot(displacement)
```

```python
dmax = []
dmean = []
for _ in range(10000):
    x = np.random.randn(1000, 2)
    position = np.cumsum(x, 0)
    displacement = np.linalg.norm(position, axis=-1)
    dmax.append(np.max(displacement))
    dmean.append(np.mean(displacement))
plt.hist(dmax, bins=50, histtype='step');
plt.hist(dmean, bins=50, histtype='step');
```

```python
plt.plot(dmax, dmean, ',')
plt.plot([0, 140], [0, 70], 'k-')
```

```python
dmax = []
for _ in range(10000):
    x = np.random.randn(1000, 2)
    position = np.cumsum(x, 0)
    displacement = np.linalg.norm(position, axis=-1)
    dmax.append(displacement[-1])
plt.hist(dmax, bins=50);
```

```python
dim=2
thit = []
for _ in range(1000):
    t = 0
    x = np.zeros(dim)
    while np.linalg.norm(x) < 10:
        x += np.random.randn(dim)
        t += 1
    thit.append(t)
plt.hist(thit, bins=50);
```
